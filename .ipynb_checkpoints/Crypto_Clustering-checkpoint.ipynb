```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
print(df_market_data.columns)
```

```python
# Use the StandardScaler() module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

# Normalize the data without dropping "coin_id" since it's already the index
scaled_data = scaler.fit_transform(df_market_data)

# Convert the scaled data back into a DataFrame
scaled_df = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns,
    index=df_market_data.index
)

# Display the first few rows of the scaled DataFrame
scaled_df.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Loop through each k-value to compute the inertia
for k in k_values:
    # Create a KMeans model using the loop counter for n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=0)
    
    # Fit the model to the scaled data
    kmeans_model.fit(scaled_df)
    
    # Append the model's inertia to the inertia list
    inertia.append(kmeans_model.inertia_)

# Display the inertia values
print(inertia)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid()
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 4
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans = KMeans(n_clusters=optimal_k, random_state=1)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans.fit_predict(scaled_df)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the scaled DataFrame
scaled_df_copy = scaled_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
scaled_df_copy["Cluster"] = clusters

# Display the copy of the scaled DataFrame
print(scaled_df_copy.head())
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scaled_df_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols=["coin_id"]
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(scaled_df_copy.drop(columns=["Cluster"]))
pca_df = pd.DataFrame(pca_data, columns=["PC1", "PC2", "PC3"], index=scaled_df.index)

# View the scaled PCA data
print(pca_df)
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
print(pca.explained_variance_ratio_)
total_variance = sum(pca.explained_variance_ratio_)
print(f"Total Explained Variance: {total_variance}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.895031657030984

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Loop through each k-value to compute the inertia
for k in k_values:
    # Create a KMeans model using the loop counter for n_clusters
    pca_kmeans_model = KMeans(n_clusters=k, random_state=0)
    
    # Fit the model to the scaled data
    pca_kmeans_model.fit(pca_df)
    
    # Append the model's inertia to the inertia list
    pca_inertia.append(pca_kmeans_model.inertia_)
```

```python
# Create a DataFrame with the data to plot the Elbow curve
pca_inertia_df = pd.DataFrame({"k": k_values, "Inertia": pca_inertia})
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(8, 6))
plt.plot(pca_inertia_df['k'], pca_inertia_df['Inertia'], marker='o', color='b', linestyle='-', linewidth=2, markersize=6)
plt.title('Elbow Curve for Optimal k', fontsize=16)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 6


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Yes

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 6
```

```python
# Fit the K-Means model using the PCA data
kmeans = KMeans(n_clusters=optimal_k, random_state=1)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
clusters = kmeans.fit_predict(pca_df)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
pca_scaled_df = pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
pca_scaled_df["Cluster"] = clusters
pca_scaled_df["coin_id"] = pca_scaled_df.index

# Display the copy of the scaled PCA DataFrame
print(pca_scaled_df.head())
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
pca_scaled_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Cluster",
    hover_cols=["coin_id"]
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
plt.figure(figsize=(10, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-', label='Original Data')
plt.plot(k_values, inertia, marker='s', linestyle='--', label='PCA Data')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
